```python
## create an empty list to the results
job_title_all = []
payment_all = []
deadline_all = []
job_description_all = []
job_category_all = []
job_owner_all = []
country_of_job_owner_all = []

urls = []
for pgs in range(2, 84):
    u1 = "https://www.guru.com/d/jobs/pg/"
    url = u1 + str(pgs) + '/'
    response = requests.get(url=url)
    freelance_soup = BeautifulSoup(response.content)
    container = jobs_soup.find_all("div", class_ = "record__details")
    for jobs in container:
            ## the postfix 1 at the at the end indicates for page 1
            job_title_1 = jobs.find('h2', class_="jobRecord__title jobRecord__title--changeVisited").text.strip()
            job_title_all.append(job_title_1)
            payment_1 = jobs.find('div', class_="jobRecord__budget").text.strip()
            payment_all.append(payment_1)
            deadline_1 = jobs.find("p", class_="copy small grey rhythmMargin1").text.strip()
            deadline_all.append(deadline_1)
            job_description_1 = jobs.find("p", class_="jobRecord__desc").text.strip()
            job_description_all.append(job_description_1)
            main_tag_1 = jobs.find("span", class_="skillsList__skill greyBackground").text.strip()
            job_category_all.append(main_tag_1)
            job_owner_1 = jobs.find('div', class_="avatarinfo").h3.text.strip()
            job_owner_all.append(job_owner_1)
            job_owner_1 = jobs.find('p', class_="subtext freelancerAvatar__subText darkGrey").text.strip()
            country_of_job_owner_all.append(job_owner_1)
```

```python
## create a data frame
freelance_jobs_dict = {'job_title' : job_title_all, 
'payment' : payment_all, 
'deadline' : deadline_all, 
'job_category' : job_category_all, 
'job_owner' : job_owner_all, 
'country_of_job_owner' : country_of_job_owner_all
}
freelance_jobs_df = pd.DataFrame(freelance_jobs_dict)
freelance_jobs_df
```

```python
## append the data from page 1
jobs_scraped = [freelance_jobs_df, jobs_df_pg1]
freelance_jobs = pd.concat(jobs_scraped)
```

The payment variable is not informative per now

Due to missing prices and durations, we can retaing only whether the price is fixed or hourly rate

Also get the actual deadline dates

```python
## retain only a label indicating whether the price is fixed or hourly
def get_label(var):
    reg = re.compile(r'(\|.*)|\n')
    after_extra = re.sub(reg, '', var)
    return after_extra


## get the actual deadline date
def get_date(var):
    reg = re.compile(r'Send before ')
    after_sub = re.sub(reg, '', var)
    return after_sub

freelance_jobs['deadline'] = freelance_jobs['deadline'].apply(get_date)
freelance_jobs['payment'] = freelance_jobs['payment'].apply(get_label)
freelance_jobs
```
